```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
s1 = 0.03; s2 = 0.04; s3 = 0.05; s4 = 0.05

# s1 = 0.05; s2 = 0.04; s3 = 0.03; s4 = 0.03

# s1 = 0.03; s2 = 0.04; s3 = 0.05; s4 = 0.03

spot = [s1, s2, s3, s4]
term = np.arange(1,5)

f1 = s1
f12 = (1+s2)**(2) / (1+s1)**(1) -1
f23 = (1+s3)**(3) / (1+s2)**(2) -1
f34 = (1+s4)**(4) / (1+s3)**(3) -1
forward = [f1, f12, f23, f34]

plt.figure(figsize=[10,8])
plt.plot(term, spot, label='Spot')
plt.plot(term, forward, label='Forward Curve')
plt.legend()
plt.show()
```
